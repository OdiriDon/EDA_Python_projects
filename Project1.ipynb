```python
#14. How Many People Have A Credit Card That Expires In 2020

df['CC Exp Date']
```

```python
def fun():
    count = 0
    for date in df['CC Exp Date']:
        if date.split('/')[1] == '20':
            count = count + 1
    return count
```

```python
fun()
```

```python
df[df['CC Exp Date'].apply(lambda x:x[3:] == '20')].count()
```

```python
#17. Replace Salary Values['<= 50k', '>= 50k'] With 0 And 1

df['income'].unique()
```

```python
df['income'].value_counts()
```

```python
sns.countplot('income', data = df)
```

```python
def  salary_data(sal):
    if sal == '<=50k':
        return 0
    else:
        return 1
```

```python
df['encoded_salary'] = df['income'].apply(salary_data)
```

```python
df['encoded_salary']
```

```python
df
```

```python
df['income'].replace(to_replace = ['<=50k', '>=50k'], value = [0, 1], inplace = True)
```

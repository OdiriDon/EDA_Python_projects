## Let's answer some questions

```python
# What are the most common crimes in terms of offense group?

df.OFFENSE_CODE_GROUP.value_counts()
```

```python
# Creating a viz for our findings above

offense_group_vals = df.OFFENSE_CODE_GROUP.value_counts()[:10]

display(offense_group_vals/df.shape[0])

((offense_group_vals/df.shape[0])*100).plot(kind='bar');
plt.title('Top 10 Offense Groups (as % of all crime)');
```

```python
# What are the least commono offense group

df.OFFENSE_CODE_GROUP.value_counts().sort_values(ascending=True)[:10]
```

```python
# What are the most common offense descriptions?

offense_des = df.OFFENSE_DESCRIPTION.value_counts()[:10]/df.shape[0]*100

(offense_des).plot(kind='bar')
plt.title('Most common offense description by % of total crime')
```

```python
# Which year were the most crimes committed?

df.groupby('YEAR').count()['INCIDENT_NUMBER'].plot(kind='bar');
plt.title('Number of crimes');
```

```python
# Are there more crimes committed on specic days let's find out

df.groupby('DAY_OF_WEEK').count()['INCIDENT_NUMBER'].plot(kind='bar')
plt.title('Days with most crimes')
```

```python
# Are there more crime in specific hours?

df.groupby('HOUR').count()['INCIDENT_NUMBER'].plot(kind='bar')
```

```python
# On what days and during which hours are the most crimes committed?

df.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER']

week_and_hour = df.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()
week_and_hour
```

```python
# On what days and during which hours are the most crimes committed?

week_and_hour.columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

week_and_hour
```

```python
sns.heatmap(week_and_hour)
```

```python
# In which month were the number of crime below average?

avg_crime = df.groupby(['YEAR','MONTH']).count()['INCIDENT_NUMBER'].mean()
avg_crime

print('The average number of crime is ' + str(avg_crime))

year_and_month = df.groupby(['YEAR','MONTH']).count()['INCIDENT_NUMBER'].unstack()
av_cr = year_and_month.where(year_and_month<avg_crime)
av_cr
```

```python
# In which districts were the most crimes committed on a yearly basis
df.groupby(['DISTRICT','YEAR']).count()['INCIDENT_NUMBER'].unstack().plot(kind='bar')
plt.title('Districts with most crimes yearly')
```
